```r
ricker <- function(nzero, r, K=1, time=100, from=0, to=time)
{
    N <- numeric(time + 1)
    N[1] <- nzero
    for(i in 1:time) N[i+1] <- N[i]*exp(r*(1 - N[i]/K))
    Time <- 0:time
    print(Time)
    plot(Time, N, type="l", xlim=c(from, to))
}
```

```r
layout(matrix(1:3, 3, 1))
```

```r
ricker(0.1, 1); title("r = 1")
ricker(0.1, 1); title("r = 1")
ricker(0.1, 3); title("r = 3")
```
